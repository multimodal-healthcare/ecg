```python
from pathlib import Path
import pandas as pd
from typing import List
```

```python
project_path = Path.cwd()
csv_dir = project_path/'data/coherent/csv'
ecg_csv = csv_dir/'ecg.csv'
train_patients_csv = csv_dir/'train_patients.csv'
val_patients_csv = csv_dir/'valid_patients.csv'
test_patients_csv = csv_dir/'test_patients.csv'
```

```python
ecg_df = pd.read_csv(ecg_csv)
train_patients_df = pd.read_csv(train_patients_csv)
val_patients_df = pd.read_csv(val_patients_csv)
test_patients_df = pd.read_csv(test_patients_csv)
```

```python
ecg_df.head(5)
```

```python
train_patients_df.head(5)
```

```python
def process_output(df: pd.DataFrame, picked_cols: List, labels: List):
    new_df = df[picked_cols]
    new_df[labels] = new_df[labels].astype(int)
    new_df['label'] = new_df[labels].values.tolist()
    new_df = new_df[['patient', 'label']]
    
    return new_df
```

```python
cols = ['patient', 'heart_failure', 'coronary_heart', 'myocardial_infarction', 'stroke', 'cardiac_arrest']
labels =  ['heart_failure', 'coronary_heart', 'myocardial_infarction', 'stroke', 'cardiac_arrest']
train_patient_label_df = process_output(train_patients_df, cols, labels)
val_patient_label_df = process_output(val_patients_df, cols, labels)
test_patient_label_df = process_output(test_patients_df, cols, labels)
```

```python
train_patient_label_df.head()
```

```python
print(f'train df len = {len(train_patient_label_df)}')
print(f'val df len = {len(val_patient_label_df)}')
print(f'test df len = {len(test_patient_label_df)}')
```

Merge ecg dataframe with output label dataframe 

```python
train_ecg_label_df = pd.merge(ecg_df, train_patient_label_df, on='patient', how='right')
train_ecg_label_df
```

```python
train_ecg_label_df = train_ecg_label_df.dropna()
train_ecg_label_df
```

```python
val_ecg_label_df = pd.merge(ecg_df, val_patient_label_df, on='patient', how='right')
test_ecg_label_df = pd.merge(ecg_df, test_patient_label_df, on='patient', how='right')
val_ecg_label_df = val_ecg_label_df.dropna()
test_ecg_label_df = test_ecg_label_df.dropna()
```

Save things to csv

```python
train_ecg_label_df.to_csv(csv_dir/'train_ecg.csv')
val_ecg_label_df.to_csv(csv_dir/'val_ecg.csv')
test_ecg_label_df.to_csv(csv_dir/'test_ecg.csv')
```
